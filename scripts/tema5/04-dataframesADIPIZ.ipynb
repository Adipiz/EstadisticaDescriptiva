# Cómo crear un data frame

```python
import numpy as np
```

```python
df = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
```

```python
df
```

```python
df.mean()  # Calcular la medias por las filas
```

```python
df.mean(axis = 'columns')  # hacer el promedio por columnas
```

```python
df2 = pd.DataFrame({
    'Key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'Values' : range(6)
}, 
columns = ['Key', 'Values']
)

df2
```

```python
df2.groupby('Key') 
```

```python
df2.groupby('Key').mean()
```

# Manipulación de data frames

```python
df = pd.DataFrame({
    'Key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'Value1': range(6),
    'Value2': np.random.randint(0,100,6)
    })
df
```

# Aggregate

```python
df.groupby('Key').aggregate([min, np.median, np.mean, max]) # queremos aplicar varias funciones 
```

```python
df.groupby('Key'). aggregate({       # aplicamos el diccionario en aggr para aplicar funciones específicas
 'Value1': 'min',
 'Value2': 'max'  
})
```

# Filter

```python
df.groupby('Key').std()
```

```python
def filter_func(x):
    return x['Value2'].std()>10
```

```python
df.groupby('Key').filter(filter_func)
```

```python
display(df, df.groupby('Key').std(),df.groupby('Key').filter(filter_func))
```

## Transform 

Se utiliza cuando queremos una versión reducida de los datos sin combinar

```python
df.groupby('Key').transform(lambda x: x - x.mean())
```

## Apply

```python
def norm_by_col(x):
    x['Value1'] = x['Value1']/x['Value2'].sum()
    return x
```

```python
df.groupby('Key').apply(norm_by_col)
```

# División de datos por columnas

```python
L = [0,1,0,1,2,0] # fíjate que es una lista que clasifica lo que hay en cada fila
print(df)
df.groupby(L).sum()
```

```python
df.groupby('Key').sum()
```

```python
df2 = df.set_index('Key')  # utilizo esta nomenclatura para fijar el nombre de las filas 
df2
```

```python
mapping = {                    # utilizamos un diccionario para agrupar
    'A': 'vowel',
    'B': 'consonant',
    'C': 'consonant'
}
```

```python
df2.groupby(mapping).sum()
```

```python
df2.groupby(str.lower).mean()
```

```python
df2.groupby([str.lower,mapping]).mean()
```

# Gráficos desde data frames y exportación

```python
years = [year for year in range(1900, 2020)] # creamos un array desde 1990 hasta 2020
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    "Year": years,
    "Deads": deads
})

df.head(3)
```

```python
# Gráfica

df.plot(kind = "scatter", x = "Year", y = "Deads", figsize =  (10,8))
```

```python
my_plot = df.plot(x = "Year", y = "Deads", figsize =  (10,8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("../../data/data_frame_plotADIPIZ.png")
```
