```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1)  # opcional si es una 
plt.subplot(3,1,1)
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k' )

plt.subplot(3,1,2)
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
plt.figure(1)  # Figura 1
plt.subplot(3,1,1)
plt.title("Figura 1")

plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k' )
plt.subplot(3,1,2)
plt.plot(x2, f(x2), 'g--')


plt.figure(2)  # Figura 2
plt.plot([4,5,6,7])
plt.title("Figura 2")
```

# Textos en los grficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, density=True, facecolor = "g", alpha = 0.3)
plt.xlabel("Cociente intelectual")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$", fontsize = 14, color = "black")
plt.text(121,0.015, r'$\mu = 100,\ \sigma=20$')

plt.axis([20,180, 0, 0.025])
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)

plt.plot(x,y,lw = 2.0)

plt.annotate('Mximo local', xy = (4*np.pi,1), xytext = (15,1.5), 
            arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.ylim(-2,2)
```

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd= 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
```

```python
plt.figure(figsize=(10,6), dpi = 120)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.ticklabel_format(useOffset=False)
plt.yscale("log")
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title("Escala Logartmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthresh = 0.01)
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title("Escala Log Simtrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.ticklabel_format(style='plain')
plt.yscale("logit")
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title("Escala logstica")
plt.grid(True)


plt.subplots_adjust(top = 0.80, bottom = 0.03, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)

plt.show()
```

# Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize =(10,8))
plt.plot(x,S, color = "blue", linewidth = 1.2, linestyle = "-", label = 'seno')
plt.plot(x,C, color = "green", linewidth = 1.2, linestyle = "-", label = 'coseno')
plt.xlim(-4,4)
plt.ylim(-1.2,1.2)
plt.ylim(S.min()*1.1, S.max()*1.1)
# plt.xticks(np.linspace(-4,4,9, endpoint = True))
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks(np.linspace(-1,1,3, endpoint = True), ['-1', '', '+1'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc = 'upper left')

x0 = 2*np.pi/3
x1 = -np.pi/3

plt.plot([x0,x0], [0, np.sin(x0)], color = 'blue', linewidth = 2.5, linestyle = '--')
plt.scatter([x0,], [np.sin(x0), ], 50, color = "blue")
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy = (x0, np.sin(x0)), xycoords = 'data',
                                                               xytext = (+10,+30), textcoords = "offset points",
            fontsize = 15, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.plot([x0,x0], [0, np.cos(x0)], color = 'green', linewidth = 2.5, linestyle = '--')
plt.scatter([x0,], [np.cos(x0),], 50, color = "green")
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy = (x0, np.cos(x0)), xycoords = 'data',
                                                               xytext = (+45,-25), textcoords = "offset points",
            fontsize = 15, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.plot([x1,x1], [0, np.cos(x1)], color = 'green', linewidth = 2.5, linestyle = '--')
plt.scatter([x1,], [np.cos(x1),], 50, color = "green")
plt.annotate(r'$\cos(-\pi/3)$', xy = (x1, np.cos(x1)), xycoords = 'data',
                                                               xytext = (-115,-25), textcoords = "offset points",
            fontsize = 15, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = "white", edgecolor  ="None", alpha = 1))

plt.show()
```
