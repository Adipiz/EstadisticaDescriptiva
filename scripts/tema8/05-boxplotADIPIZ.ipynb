# Diagrama de cajas y bigotes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19880519)
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 50*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)
data = np.concatenate((dist1,dist2,dist3,dist4))
```

```python
fig1, ax1=plt.subplots()
ax1.set_title("boxplot básico")
ax1.boxplot(data)
```

```python
fig2, ax2=plt.subplots()
ax2.set_title("boxplot con notch")
ax2.boxplot(data, notch = True)
```

```python
green_diamond = dict(markerfacecolor="g",marker ="d")
fig3, ax3=plt.subplots()
ax3.set_title("boxplot con notch y outliers personalizados")
ax3.boxplot(data, notch = True, flierprops=green_diamond)
```

```python
fig4, ax4=plt.subplots()
ax4.set_title("boxplot con notch y sin outliers")
ax4.boxplot(data, notch = True, showfliers = False)
```

```python
red_square = dict(markerfacecolor="r", marker="s")
fig5, ax5=plt.subplots()
ax5.set_title("boxplot con notch y en horizontal")
ax5.boxplot(data, vert = False, flierprops=red_square, whis = 1.5) # es el parámetro que fija el IQR
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 40*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)

data2 = np.concatenate((dist1,dist2,dist3,dist4))
```

```python
data.shape = (-1,1)
data2.shape = (-1,1)
```

```python
fulldata = [data.flatten(), data2.flatten(), data2[::2].flatten()]
fulldata
```

```python
fig7, ax7 = plt.subplots()
ax7.set_title("Múltiples muestras de diferentes dist")
ax7.boxplot(fulldata, notch = True)
plt.show()
```
